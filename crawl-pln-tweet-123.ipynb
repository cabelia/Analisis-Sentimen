```python
import nest_asyncio
nest_asyncio.apply() 
import twint
import pandas as pd

c = twint.Config()
c.Since = "2022-01-01"
c.Until = "2022-07-01"
c.Lang = "id"
c.Limit = 300
c.Search = 'pln_123'
c.Hide_output = True
c.Debug = True
c.Pandas = True
```

```python
%%time
try:
    print("Start Crawling")
    
    def twint_to_dataframe(colums):
        return twint.output.panda.Tweets_df[colums]
    
    #start crawling
    twint.run.Search(c)
    
    data = twint_to_dataframe(['date','username','tweet'])
    
except Exception as e:
    print(e)
    
print("Finished")
```
